```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data_path = r'C:\Users\Habiba\Machine learning\Machine-Learning-Course-main 2\Machine-Learning-Course-main\Data\smoke_detection_iot.csv'
df = pd.read_csv(data_path)

# Count the number of instances of each label
counts = df['Fire_Alarm'].value_counts()

# Plot the counts as a pie chart
plt.pie(counts, autopct='%1.1f%%')
plt.title('Fire Alarm graph')
plt.legend(['No_Fire','Yes_Fire'], loc='lower right')
plt.show()
```
